```
import numpy as np
import matplotlib.pyplot as plt

#exemplos?

```

1) $y=x^2$  
1.1) $y2=x^3$


```
# minha proposta de grafico
x = np.linspace(5, 25, 20)
y= np.power(x,2)
plt.figure(dpi=100)
plt.plot(x,y,'c-.',linewidth= 3)
plt.xticks(rotation=70)
plt.yticks(rotation='vertical')
plt.legend('gráfico')
plt.grid(color='m', linestyle='-.', linewidth=0.3)
```

```
x= np.linspace(-10,10,50)
y= np.power(x,2)
y2= np.power(x,3)
# Acabei de criar as coordenadas
plt.figure(dpi=150, facecolor = 'c')
plt.plot(x,y,'k-',label= '$y=x^2$',linewidth= 2)
plt.plot(x,y2,'r-',label= '$y2=x^3$',linewidth= 3)
plt.xticks(rotation=85)
plt.yticks(rotation=35)
plt.legend()
plt.grid(color='g', linestyle='-.', linewidth=1)
```

2) $y=sin(x)$


```
x= np.linspace(0, 2* np.pi ,40)
y= np.sin(x)

plt.figure()
plt.plot(x,y,'bh',label='$y=sin(x)$',linewidth= 1.1)
plt.legend()
```

3) $y=x^2 -5x +6$


```
x= np.linspace(-100,100,1000)
y= np.power(x,2) - np.multiply(x,5) + 6

plt.figure()
plt.plot(x,y,'g-',label='$y=x^2 -5x +6$',linewidth= 1.2)
plt.legend()
```

4) $f_1(x)=\frac{x}{1+x^2}$




```
x= np.linspace(0,100,1000)
y= x/(1 + np.power(x,2))

plt.figure()
plt.plot(x,y,'c-',label='$f_1(x)$',linewidth= 1.0)
plt.legend(loc='best')
```

onde $f_1(x)= \frac{x}{1+x^2}$ representa a funcao fualana que estou desenhando...
