**Functions(Funções)**

As functions são definidas em python por meio da palavra chava *def* seguida do nome e dos eventuais argumentos em entrada entre parentesis. As linha de definição tem que terminar com (:)

As functions podem devolver um ou mais valores por meio da palavra chave *return*


```
def square(x):
    return x**2
```

```
# Uso como variavel
x = 2
y = square(x)
print(y)
```

```
y
```

```
# uso imediato tipo calculadora
square(125)
```

```
z = square(125)
```

**ESTRUTURA DO *IF ELSE* em PYTHON**

image.png

```
def valorAbsoluto2(numero):
  if (numero<0) :
    r=numero*-1
    print('passei por aqui sem else')
  else :
    r=numero
    print('passei por aqui no else')
  return r
```

```
print(valorAbsoluto2(-5))
```

```
print(valorAbsoluto2(5))
```

```
def numeroPar(numero):
  if (numero%2 == 0):
    print('O numero é par')
  else :
    print('O numero é impar')
```

```
numeroPar(444)
```

```
numeroPar(555)
```

**ESTRUTURA DO *FOR IN N* em PYTHON**


FOR Pode ser usado para iterar sequencias como list, string, tuple, set, array, data frame.

Para iterar uma serie de itens o FOR usa a função RANGE que devolve uma lista numerada de uma sequencia.

Os FOR in Python são zero-indexed.


```
print(range(2,10))
```

```
for contadorDeRepetição in range(5):
    print('contadorDeRepetição = '+str(contadorDeRepetição)+' Oi tudo bem?')
    print('Semana 6')
print('Terminei')
```

```
for i in range(10):
    print (i)
```

```
sequence = [1,2,8,100,200,'caipira','caiçãra']

for i in sequence:
    print (i)
```



**Lembre-se: RANGE sempre deve usar numeros inteiros.**

```
len(sequence)
```

```

for i in range(len(sequence)):
    print (sequence[i])

```

```
# RANGE(Incio,  Fim, Passo)
for i in range(10,90,7):
    print (i)
```
