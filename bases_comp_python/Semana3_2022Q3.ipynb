```
#https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html
import numpy as np
print(np.random.randint(0,10, size=3))
```

```

vals = np.random.randint(0,10,size=1)
more_vals = np.sqrt(np.random.randint(0,10,size=5))

print(vals)
print(more_vals)
```

```
more_vals_ceil = np.ceil(more_vals)
print(more_vals_ceil)
```

```
more_vals_floor = np.floor(more_vals)
print(more_vals_floor)
```

```
novo_conjunto = np.power(more_vals_floor,2)
print(novo_conjunto)
```

```
novo_conjunto = novo_conjunto*15
print(novo_conjunto)
```
